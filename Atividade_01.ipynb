```python
# QUESTÃO 01 

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Conjunto A
A = pd.DataFrame({
    'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y': [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
})

# Regressão linear
regressao_A = LinearRegression().fit(A[['x']], A['y'])
print(f"Correlação A: {A.corr()['y'][0]}")
print(f"Reta de regressão A: y = {regressao_A.intercept_:.2f} + {regressao_A.coef_[0]:.2f}x")
print(f"Predição A : {regressao_A.predict([[10]])[0]:.2f}")

# Gráfico 
plt.scatter(A['x'], A['y'])
plt.plot(A['x'], regressao_A.predict(A[['x']]), color='blue')
plt.title('Conjunto A')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Conjunto B
B = pd.DataFrame({
    'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y': [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
})

regressao_B = LinearRegression().fit(B[['x']], B['y'])
print(f"Correlação B: {B.corr()['y'][0]}")
print(f"Reta de regressão B: y = {regressao_B.intercept_:.2f} + {regressao_B.coef_[0]:.2f}x")
print(f"Predição B : {regressao_B.predict([[10]])[0]:.2f}")

# Gráfico 
plt.scatter(B['x'], B['y'])
plt.plot(B['x'], regressao_B.predict(B[['x']]), color='blue')
plt.title('Conjunto B')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Conjunto  C
C = pd.DataFrame({
    'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y': [7.46, 8.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
    })

regressao_C = LinearRegression().fit(C[['x']], C['y'])
print(f"Correlação C: {C.corr()['y'][0]}")
print(f"Reta de regressão C: y = {regressao_C.intercept_:.2f} + {regressao_C.coef_[0]:.2f}x")
print(f"Predição C : {regressao_C.predict([[10]])[0]:.2f}")

# Gráfico 
plt.scatter(C['x'], C['y'])
plt.plot(C['x'], regressao_C.predict(B[['x']]), color='blue')
plt.title('Conjunto C')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Conjunto D
D = pd.DataFrame({
    'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y': [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 5.56, 7.91, 6.89, 12.50]
    })

regressao_D = LinearRegression().fit(D[['x']], D['y'])
print(f"Correlação D: {D.corr()['y'][0]}")
print(f"Reta de regressão D: y = {regressao_D.intercept_:.2f} + {regressao_D.coef_[0]:.2f}x")
print(f"Predição D : {regressao_D.predict([[10]])[0]:.2f}")

# Gráfico 
plt.scatter(D['x'], D['y'])
plt.plot(D['x'], regressao_D.predict(D[['x']]), color='blue')
plt.title('Conjunto D')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#QUESTÃO 03
#A)

import matplotlib.pyplot as plt

sexo = ['M', 'M', 'F', 'F', 'F', 'F', 'M', 'F', 'F', 'M']
massa = [62, 62.9, 36.1, 54.6, 48.5, 42, 47.4, 50.6, 42, 48.7]
taxa = [1792, 1666, 995, 1425, 1396, 1418, 1362, 1502, 1256, 1614]

fig, ax = plt.subplots()
colors = {'M':'green', 'F':'pink'}
ax.scatter(massa, taxa, c=[colors[s] for s in sexo])
ax.set_xlabel('Massa do corpo sem gordura')
ax.set_ylabel('Metabolismo')
plt.show()



#B)
import numpy as np

x = np.array(massa)
y = np.array(taxa)

# cálculo dos coeficientes da reta de regressão
beta_1, beta_0 = np.polyfit(x, y, 1)

# plotagem do diagrama de dispersão e da reta de regressão
fig, ax = plt.subplots()
ax.scatter(massa, taxa, c=[colors[s] for s in sexo])
ax.plot(x, beta_0 + beta_1*x, color='blue')
ax.set_xlabel('Massa do corpo sem gordura')
ax.set_ylabel('Metabolismo')
plt.show()

#C)

print("O intercepto tem valor de 1212,69 , mas avaliá-lo nesse caso nao tem um sentido abrangente , pois nao há pessoas que pesam zero.Já a inclinação de 22,62 mostra o aumento do metabolismo para cada aumento de 1 kg na massa do corpo sem gordura.")



#D) 
homem = (np.array(sexo) == 'M')
mulher = (np.array(sexo) == 'F')
corr_homem = np.corrcoef(x[homem], y[homem])[0, 1]
corr_mulher = np.corrcoef(x[mulher], y[mulher])[0, 1]
print("Coeficiente de Pearson / homem:", corr_homem)
print("Coeficiente de Pearson / mulher:", corr_mulher)

```

```python
import matplotlib.pyplot as plt
import numpy as np

idade = np.array([71, 64, 43, 67, 56, 73, 68, 56, 76, 65, 45, 58, 45, 53, 49, 78, 73, 68])
massa = np.array([82, 91, 100, 68, 87, 73, 78, 80, 65, 84, 116, 76, 97, 100, 105, 77, 73, 78])

# diagrama de dispersão
plt.scatter(idade, massa)
plt.title('Massa Muscular x Idade')
plt.xlabel('Idade')
plt.ylabel('Massa Muscular')
plt.show()

# correlação linear
coeficiente_linear= np.corrcoef(idade, massa)[0, 1]
print('Coeficiente Linear:', coeficiente_linear)

# reta de regressão
a, b = np.polyfit(idade, massa, 1)
print('a =', a, 'b =', b)

plt.scatter(idade, massa)
plt.plot(idade, a*idade + b, color='red')
plt.title('Massa Muscular x Idade')
plt.xlabel('Idade')
plt.ylabel('Massa Muscular')
plt.show()

# IDADE = 50
idade_est = 50
massa_est= a*idade_est + b
print('Massa Muscular estimada =', idade_est, ':', massa_est)
```

```python
#QUESTAO 05
#A)
import matplotlib.pyplot as plt
import numpy as np

T = np.array([8, 9, 7, 8, 6, 8, 5, 5, 6, 7, 4, 7, 3, 5, 3])
V = np.array([14, 13, 12, 13, 10, 12, 11, 11, 10, 12, 10, 13, 20, 12, 11])

plt.scatter(T, V)
plt.xlabel('T')
plt.ylabel('V(em milhares)')
plt.show()


#B)

corr = np.corrcoef(T, V)[0, 1]
print("Correlação:", corr)

#C)
a = np.cov(T, V)[0, 1] / np.var(T)
b = np.mean(V) - a * np.mean(T)

T_lin = np.linspace(T.min(), T.max(), 100)
V_lin = a * T_lin + b

# Gráfico
plt.scatter(T, V)
plt.plot(T_lin, V_lin, color='red')
plt.xlabel('T')
plt.ylabel('V')
plt.title('Diagrama de Dispersão e Reta de Regressão')
plt.show()


#D) 
print('O valor de 20 na variável V é um outlier ')
```

```python
#QUESTAO 06 

#A)
import matplotlib.pyplot as plt

diametro = [22.5, 28, 27.5, 25.5, 22, 30.5, 23, 25, 23.5, 27, 21.5, 22, 29, 20.5, 27]
log = [0.19, 0.62, 0.51, 0.53, 0.24, 0.87, 0.25, 0.25, 0.37, 0.32, 0.13, 0.35, 0.53, 0.22, 0.65]

plt.scatter(diametro, log)
plt.xlabel("Diâmetro ")
plt.ylabel("Log")
plt.show()

#B)

diam_mean = np.mean(diametro)
log_mean = np.mean(log)

diam_log_sum = np.sum((diametro - diam_mean) * (log - log_mean))
diam_sum = np.sum((diametro - diam_mean)**2)

b = diam_log_sum / diam_sum
a = log_mean - b * diam_mean

print("Coeficiente angular:", b)
print("Coeficiente linear:", a)

# Convertendo a lista para um array do NumPy
diam_array = np.array(diametro)
conversao = b * diam_array

print(conversao)

#Gráfico
plt.scatter(diametro, log)
plt.plot(diametro, a + conversao , color='blue')
plt.show()

#C) 
corr = np.corrcoef(diametro, np.log10(log))[0, 1]
print("Correlação:", corr)
print("Correlação forte e positiva entre o diâmetro e o log")

#D)
print("Porque os pontos não estão bem distribuídos ao redor da reta.")

#E)
import statsmodels.api as sm
from statsmodels.formula.api import ols

dados = {"Diametro": diametro, "log": np.log10(log)}
df = pd.DataFrame(dados)

modelo = ols('diametro ~ log', data=df).fit()
anova = sm.stats.anova_lm(modelo, typ=2)
print(anova)





```
